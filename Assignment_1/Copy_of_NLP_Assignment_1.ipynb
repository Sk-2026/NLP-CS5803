```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
domain = "https://www.wikipedia.org/wiki"
query = "Tourism_in_Telangana"
URL = f"{domain}/{query}"

response = requests.get(URL)
if response.status_code == 200:
    print("Successfully fetched webpage!")
else:
    print("Failed to fetch the webpage. Status code:", response.status_code)
```

```python
# Create a BeautifulSoup object and extract specific content from the HTML page.
soup = bs(response.content, 'html.parser')

# Pretty print the HTML content
print(soup.body.prettify()) # soup.body.prettify() also works
```

```python
links = soup.find_all("a")
links = [link['href'] for link in links if 'href' in link.attrs]
links
```

```python
# Filter out links that are not in the 'en.wikipedia.org' domain and exclude certain domains
english_links = [link for link in links if "https://" not in link]
english_links = list(set(english_links))
english_links
```

```python
from urllib.parse import urljoin
links = [urljoin(domain, link) for link in english_links]
links
```

```python
len(links)
```

```python
data = []

for link in links:
    try:
        response = requests.get(link)
        soup = bs(response.content, 'html.parser')

        # Extract title
        title = soup.find('title').get_text(strip=True) if soup.find('title') else "No Title"

        # Extract page text
        text = soup.get_text(separator=' ', strip=True)

        # Append to data
        data.append([title, link, text])
        print(f"Scraped: {title} | URL: {link}")
    except Exception as e:
        print(f"Error scraping {link}: {e}")
```

```python
# A sample document
for i in range(3):
  print(data[0][i])
```

```python
import re
import nltk
from nltk.corpus import stopwords

# Download the stopwords dataset
nltk.download('stopwords')

# Define the stop words set
stop_words = set(stopwords.words('english'))

def clean_text(text):
    # Remove special characters and numbers using regex
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = re.sub(r'\d+', '', text)

    # Tokenize the text into words
    words = text.split()

    # Remove stop words and convert to lowercase
    words = [word.lower() for word in words if word.lower() not in stop_words]

    # Join the cleaned words back into a single string
    return ' '.join(words)

# Clean each document in the dataset
for doc in data:
    doc[2] = clean_text(doc[2])  # Clean the text in the third field of each document

# Print cleaned text for the first document
print("Title:", data[0][0])
print("URL:", data[0][1])
print("Cleaned Text:", data[0][2][:500])  # Displaying only the first 500 characters
```

```python
import spacy

# Load spaCy's English language model
nlp = spacy.load('en_core_web_sm')

# Tokenization function
def tokenize_text_spacy(text):
    doc = nlp(text)

    # Sentence Tokenization
    sentences = [sent.text for sent in doc.sents]

    # Word Tokenization
    words = [token.text for token in doc]

    return sentences, words

# Apply tokenization to each document
for doc in data:
    sentences, words = tokenize_text_spacy(doc[2])  # Tokenize the cleaned text
    doc.append({'sentences': sentences, 'words': words})  # Append tokens to the document

# Print tokenized output for the first document
print("Title:", data[0][0])
print("URL:", data[0][1])
print("Sentences:", data[0][3]['sentences'][:3])  # Displaying the first 3 sentences
print("Words:", data[0][3]['words'][:20])  # Displaying the first 20 words
```

```python
from nltk.stem import PorterStemmer

# Initialize the Porter Stemmer
stemmer = PorterStemmer()

# Stemming function
def apply_stemming(words):
    return [stemmer.stem(word) for word in words]

# Apply stemming to each document
for doc in data:
    words = doc[3]['words']
    stemmed_words = apply_stemming(words)
    doc[3]['stemmed_words'] = stemmed_words  # Append stemmed words to the document

# Print the stemmed words for the first document
print("Title:", data[0][0])
print("Stemmed Words:", data[0][3]['stemmed_words'][:20])  # Displaying the first 20 stemmed words
```
